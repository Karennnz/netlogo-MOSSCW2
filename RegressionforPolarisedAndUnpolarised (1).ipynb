```python
import pandas as pd
# File paths for the provided data
non_polarized_data_path = 'abm_milk_choice_upload experimentNonPolarizedNew-table .csv'
polarized_data_path = 'PolarizedNew-table.csv'
# Load the dataframes
polarized_df = pd.read_csv(polarized_data_path, skiprows=6, sep=",")
non_polarized_df = pd.read_csv(non_polarized_data_path, skiprows=6, sep=",")
non_polarized_df.info()
```

```python
polarized_df.head()
```

## for polarized data

```python
# Filter for the last step of each run number
filtered_data =polarized_df.loc[polarized_df.groupby("[run number]")["[step]"].idxmax()]

# Filter for the first step of each run number
first_step_data = polarized_df.loc[polarized_df.groupby("[run number]")["[step]"].idxmin()]

# Merge the filtered data to calculate differences
polar_result = filtered_data.merge(
    first_step_data[["[run number]", "mean-oat", "mean-almond", "mean-soy", "mean-incum"]],
    on="[run number]",
    suffixes=("_last", "_first")
)

# Calculate the differences
polar_result["mean-oat-diff"] = polar_result["mean-oat_last"] - polar_result["mean-oat_first"]
polar_result["mean-almond-diff"] = polar_result["mean-almond_last"] - polar_result["mean-almond_first"]
polar_result["mean-soy-diff"] = polar_result["mean-soy_last"] - polar_result["mean-soy_first"]
polar_result["mean-incum-diff"] = polar_result["mean-incum_last"] - polar_result["mean-incum_first"]



# Display the resulting dataset
print(polar_result.info())
polar_result.head()
```

## scale and ols regression for polarise data

```python
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
import pandas as pd

# Assuming 'polar_result' is the DataFrame already prepared with necessary columns
# Selecting relevant columns
variables = ['network-parameter', 'social-blindness']
dependent_vars = ['mean-oat-diff', 'mean-almond-diff', 'mean-soy-diff', 'mean-incum-diff']

# Scaling independent variables
scaler = StandardScaler()
polar_result_scaled = polar_result.copy()
polar_result_scaled[variables] = scaler.fit_transform(polar_result[variables])

# Performing regression for each dependent variable
regression_results = {}
for dependent_var in dependent_vars:
    X = polar_result_scaled[variables]
    y = polar_result_scaled[dependent_var]
    X = sm.add_constant(X)  # Adding constant for intercept
    model = sm.OLS(y, X).fit()
    regression_results[dependent_var] = model.summary()

regression_results
```

## download  merged data to local

```python
import pandas as pd
# File paths for the provided data
non_polarized_data_path = 'abm_milk_choice_upload experimentNonPolarizedNew-table .csv'
polarized_data_path = 'PolarizedNew-table.csv'
# Load the dataframes
polarized_df = pd.read_csv(polarized_data_path, skiprows=6, sep=",")
non_polarized_df = pd.read_csv(non_polarized_data_path, skiprows=6, sep=",")
non_polarized_df.info()
```

```python
import pandas as pd

# Assuming the two datasets are loaded as polarized_df and non_polarized_df
# Reindexing `run number` for non_polarized and polar datasets to ensure no overlap

# Shift non_polarized run numbers to prevent overlap with polarized
non_polarized_df["[run number]"] += polarized_df["[run number]"].max() + 1

# Concatenate the two datasets
merged_df = pd.concat([polarized_df, non_polarized_df], ignore_index=True)

# Sort the merged dataframe by `run number` and `step`
merged_df = merged_df.sort_values(by=["[run number]", "[step]"]).reset_index(drop=True)

# Display the structure of the merged dataframe
merged_df.info()
merged_df.head()
```

```python
# Save the merged dataset to a CSV file
file_path = "merged_dataset.csv"
merged_df.to_csv(file_path, index=False)

```
