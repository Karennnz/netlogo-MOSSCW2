```python
# List of file names to process
file_names = [
    "abm_milk_choice_upload experimentNonPolarizedNew-table .csv", "PolarizedNew-table.csv"
]

# Initialize an empty list to store cleaned dataframes
cleaned_dfs = []

# Process each file
for file in file_names:
    # Read the dataset, skipping the first 6 rows
    df = pd.read_csv(file, skiprows=6, sep=",")  # Ensure correct delimiter (adjust to actual file format)
    
    # Check if the dataset was misread as a single column
    if len(df.columns) == 1:
        # Re-read using a different delimiter if necessary
        df = pd.read_csv(file, skiprows=6, sep="\t")
    
    # Standardize column names by stripping whitespace
    df.columns = df.columns.str.strip()
    
    # Append to the cleaned dataframes list
    cleaned_dfs.append(df)

# Align columns by filling missing columns in each dataset with NaN
all_columns = set().union(*[df.columns for df in cleaned_dfs])
aligned_dfs = [df.reindex(columns=all_columns) for df in cleaned_dfs]

# Concatenate all aligned dataframes into one long dataframe
merged_data = pd.concat(aligned_dfs, axis=0, ignore_index=True)

# Display the structure of the merged dataset
print(merged_data.info())
merged_data.head()
```

```python
grouped_data = merged_data.groupby(["[run number]", "[step]"]).mean().reset_index()
```

```python
grouped_data
```

```python
# 替换列名中的连字符
grouped_data.columns = grouped_data.columns.str.replace("-", "_")
# 确保你的变量名符合 Python 和 patsy 的要求
print(grouped_data.columns)
```

```python
from statsmodels.formula.api import mixedlm
from patsy import dmatrix

dependent_variables = ["mean_oat", "mean_almond", "mean_soy", "mean_incum"]

results = {}

for dep_var in dependent_variables:
   
    formula = f"Q('{dep_var}') ~ Q('[step]')"
    
  
    model = mixedlm(formula, grouped_data, groups=grouped_data["social_blindness"])
    result = model.fit(method='cg')  
    results[dep_var] = result.summary()


for dep_var, summary in results.items():
    print(f"Mixed Linear Model Results for {dep_var}:")
    print(summary)
    print("\n")
```

```python
from statsmodels.formula.api import mixedlm

# 定义包含交互项的模型公式
formula = "Q('mean_oat') ~ Q('[step]') + Q('social_blindness') + Q('[step]'):Q('social_blindness')"

# 拟合模型
model = mixedlm(formula, grouped_data, groups=grouped_data["[run number]"])
result = model.fit(method='cg')

# 输出结果
print(result.summary())
```

## scale data

```python
from sklearn.preprocessing import StandardScaler
from statsmodels.formula.api import mixedlm

# Step 1: 标准化数据
scaler = StandardScaler()
scaled_data = grouped_data.copy()
scaled_columns = ["[step]", "social_blindness"]
scaled_data[scaled_columns] = scaler.fit_transform(scaled_data[scaled_columns])

# Step 2: 定义模型并拟合
dependent_variables = ["mean_oat", "mean_almond", "mean_soy", "mean_incum"]
results = {}

for dep_var in dependent_variables:
    formula = f"Q('{dep_var}') ~ Q('[step]') + Q('social_blindness') + Q('[step]'):Q('social_blindness')"
    model = mixedlm(formula, scaled_data, groups=scaled_data["[run number]"])
    result = model.fit(method='cg')
    results[dep_var] = result.summary()

# Step 3: 打印结果
for dep_var, summary in results.items():
    print(f"Mixed Linear Model Results for {dep_var}:")
    print(summary)
    print("\n")
```
